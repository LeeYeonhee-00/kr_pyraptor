```python
import pandas as pd
import os
```

```python
file_path = "kr_gtfs"
files = os.listdir(file_path)
```

```python
for item in files:
    data_frame = pd.read_csv(f'./kr_gtfs/{item}', delimiter=',', encoding='utf-8', header=None, low_memory=False)
    # 첫 번째 행을 열(column)으로 설정
    data_frame.columns = data_frame.iloc[0]

    # 첫 번째 행은 더 이상 데이터로 사용하지 않으므로 삭제
    data_frame = data_frame[1:]

    print(item)
    display(data_frame)
    print()
```
